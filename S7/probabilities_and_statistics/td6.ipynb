```python
import sklearn
from sklearn.datasets import fetch_olivetti_faces
```

```python
faces = fetch_olivetti_faces()["data"]
faces_centered = faces - faces.mean(axis=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
r,c = 5,7
nc = r*c
shape = (64,64)

def plot_gallery(title,images):
    plt.figure(figsize=(2.*c,2.26*r))
    plt.suptitle(title,size=16)
    for i,comp in enumerate(images):
        plt.subplot(r,c,i+1)
        comp = comp.reshape(shape)
        vmax = comp.max()
        vmin = comp.min()
        plt.imshow(comp,cmap=plt.cm.gray,vmax=vmax,vmin=vmin)
        plt.xticks(())
        plt.yticks(())
        
    plt.subplots_adjust(0.01,0.05,0.99,0.93,0.04,0.)
plot_gallery("First centered Olivetti faces", faces_centered[:nc])
```

```python
from sklearn.decomposition import PCA
pca = PCA(20)
pca.fit(faces_centered)
```

```python
pca.explained_variance_ratio_
```

```python
comp_pca = pca.components_
plot_gallery("Plot of PCA components",comp_pca)
```
