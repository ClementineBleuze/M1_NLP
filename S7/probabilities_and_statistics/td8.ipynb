```python
data["Date"] = pd.to_datetime(data["Date"])
data
```

```python
for country in pd.unique(data["Country/Region"]):
    print(country)
```

```python
from collections import Counter
date_counter = Counter(data["Country/Region"])
print(date_counter)
```

```python
selected = [country for country in date_counter.keys() if date_counter[country] > 500]
df_covid_filtered = data[data["Country/Region"].isin(selected)]
df_covid_filtered
```

```python
df_covid_pivot = df_covid_filtered.reset_index().pivot_table(index="Date", columns='Country/Region', values='Confirmed',aggfunc = 'sum')
```

```python
df_covid_pivot=df_covid_pivot.T
```
